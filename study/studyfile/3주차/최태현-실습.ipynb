## 멜론

```python
import requests
from bs4 import BeautifulSoup
request_headers = {
    'User-Agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
)
}
with requests.get('https://www.melon.com/chart/index.htm', headers=request_headers) as res:
    soup= BeautifulSoup(res.text)
soup
```

```python
tag_list = soup.select('table div.wrap_song_info a[href*=playSong]')

for index, tag in enumerate(tag_list, 1):
    print(index, tag.string)
```

## 영화 랭킹

```python
import requests
from bs4 import BeautifulSoup
with requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&date=20201216') as res:
    soup= BeautifulSoup(res.text)
soup
```

```python
tags=soup.select('div[class=tit3]')
for i in range (len(tags)):
    print(i+1, tags[i].text.strip())
```
