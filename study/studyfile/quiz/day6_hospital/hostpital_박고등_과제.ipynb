```python
import csv 


def alllist():
    csvfile=open("hospital_2019.csv","r",encoding="utf-8")
    lines = csv.reader(csvfile)
    header= next(lines)
    for line in lines:
        print(line)
    return 0

def subject_select():
    csvfile=open("hospital_2019.csv","r",encoding="utf-8")
    lines = csv.reader(csvfile)
    header= next(lines)
    subject=[]
    lined=[]
    for line in lines:
        subject.append(line[1])
        lined.append(line)
    print("선택가능하신 과목 목록 \n",set(subject))
    subject_input=input()
    for line in lined:
        if line[1]==subject_input:
            print(line)
    return 0

def city_select():
    csvfile=open("hospital_2019.csv","r",encoding="utf-8")
    lines = csv.reader(csvfile)
    header= next(lines)
    lined=[]
    city=[]
    for line in lines:
        city.append(line[2])
        lined.append(line)
    print("선택 가능한 도시",set(city))
    city_input=input("도시를 선택하여주세요")
    for line in lined:
        if line[2]==city_input:
            print(line)
    return 0

def hospital_data(): 
    a=input("1.전체,2.과목,3.시도병원,그 이외의 입력 종료")
    
    if a=="1":
        alllist() 
        hospital_data()
    elif a=="2":
        subject_select()
        hospital_data()
    elif a=="3":
        city_select()
        hospital_data()
    else:
        print("종료되었습니다")
    return 0

    
hospital_data()
```
