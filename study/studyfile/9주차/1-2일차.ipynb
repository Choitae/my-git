### 딥러닝을 위한 기초 수학
n $\cdot$ n  행렬인 $\quad x $.
$$x=
\begin{bmatrix}
x_{00} & x_{01} & \cdots & x_{0n}\\
x_{10} & x_{11} & \cdots & x_{1n}\\
\vdots \\
x_{n0} & x_{n1} & \cdots & x_{nn}\\ 
\end{bmatrix}
$$
----
행렬은 사칙연산이 가능하다.
$$ a \cdot x = 
\begin{bmatrix}
a \cdot x_{00} & a \cdot x_{01} & \cdots & a \cdot x_{0n}\\
a \cdot x_{10} & a \cdot x_{11} & \cdots & a \cdot x_{1n}\\
\vdots \\
a \cdot x_{n0} & a \cdot x_{n1} & \cdots & a \cdot x_{nn}\\
\end{bmatrix}
$$

----
n $\cdot$ m 행렬인$\quad$y
$$y = 
\begin{bmatrix}
y_{00} & y_{01} & \cdots & y_{0m}\\
y_{10} & y_{11} & \cdots & y_{1m}\\
\vdots & \ddots \\
y_{n0} & y_{n1} & \cdots & y_{nm}\\
\end{bmatrix}
$$

x $\cdot$ y = (n$\cdot$n) $\cdot ($n$\cdot$m) = n$\cdot$m
$$ x \cdot y = 
\begin{bmatrix}
\begin{aligned}
\sum_{k=1}^nx_{0k}y_{k0} \cdots \cdots \sum_{k=0}^nx_{0k}y_{km}\\
\ddots \vdots 
\end{aligned}
\end{bmatrix}
$$



기울기 = $f(x + \Delta x) - f(x) \over \Delta x $

미분(순간 변화율) : 원 함수 $f(x)$에서 x가 a만큼 변화할 때 y는 얼마만큼 변화하는가를 알기 위함


$
f(x) \to f'(x)
$

$\begin{aligned}f'(x) = \lim_{\Delta x \to 0}{f(x + \Delta x) - f(x) \over \Delta x}\end{aligned}$ 

$f'(x) = {d \over dx}f(x) = {df(x) \over dx} = {dy \over dx}$

## 합성함수

$ u = f(x) = 2x$ 이고, 여기에서 나온 $u$값을 

$ y = g(x) = 3x$ 의 $x$값으로 대입한다고 하면,

$ y = g(f(x)) = g(x)\cdot f(x) $ 로 표기하며, 이를 $\mathbf{합성함수}$ 라고 한다  

  ----


또한 이를 미분하면,

$y' = g'(x) \cdot f'(x) = 2 \cdot 3 = 6$

즉, 각각을 미분하여 곱해주면 된다.( $\mathbf{chain Rule}$ )

${dy \over dx} = {dy \over du} \cdot {du \over dx}$

## sigmoid
   1)   $y=$ $ {1}\over {1 + e^{-e}}$

   - $y' = (1-y)y$

```python
import numpy as np
import matplotlib.pylab as plt

def sigmoid_function(x) :
    return 1 / (1+np.exp(-x))

x = np.linspace(-5,5)
y = sigmoid_function(x)

plt.plot(x,y)
plt.show()
```

  2) tanh
   - $y$ = ${e^x = e^{-x}} \over {e^x + e^{-x}}$

```python
def tanh_function(x) :
    return np.tanh(x)

x = np.linspace(-5,5)
y = tanh_function(x)

plt.plot(x, y)
plt.show()
```

  3) ReLU(Rectified Lineaer Unit)
   - $y$ = $\begin{cases}
      0\quad(x \le 0)\\
      x\quad(x > 0)
      \end{cases}$

```python
def relu_function(x):
    return np.where(x<=0, 0, x)

x = np.linspace(-5,5)
y = relu_function(x)

plt.plot(x, y)
plt.show()
```

  4) Leaky ReLU
   - x가 음수인 영역에서 아주 작은 기울기를 생성
   - 출력이 0이 되어 더이상 학습이 진행되지 않는 뉴런이 발생하는 dying ReLU 현상을 피하기 위해 사용 
   
   - $y$ = $\begin{cases}
      0.01\cdot x \quad(x \le 0)\\
      x\quad(x > 0)
      \end{cases}$

```python
def leaky_relu_function(x):
    return np.where(x <=0, 0.01*x, x)

x = np.linspace(-5,5)
y = leaky_relu_function(x)

plt.plot(x,y)
plt.show
```

  5) Softmax
   - 입력받은 값을 출력으로 0~1 사이의 값으로 모두 정규화하며 출력 값들의 총합은 항상 1이 되는 특성을 가진 함수
   - 데이터가 특정 클래스에 속할 확률이 얼마인지 구하는 함수
   - 정답이 여러 개인 데이터를 다룰 수 있다.
   - 함수가 진행되면서 normalization(정규화)를 진행하게 되고, 서로 다른 확률을 가졌지만, 모두 비슷한 영향력을 행사할 수 있도록 만들어줌.
   - 신경망의 맨 마지막 계층에서 사용
$$
f(\vec{x})_i = \begin{aligned}\frac{e^{x_i}}{\sum_{n=1}^K{e^{x_n}}}\end{aligned}
\quad for\quad i = 1, \cdots, K
$$

```python
def softmax_function(x):
    return np.exp(x)/np.sum(np.exp(x)) # 소프트맥스 함수

y = softmax_function(np.array([1,2,3]))
print(y)
```

# 인공 신경망





  - 입력에 가중치를 가하고, 이를 더해주고 $bias$ 라는 값을 더해준다.
    - $bias$ : $Activation func$를 통과 여부에 영향을 미치며, 기하학적으로는 y = ax + b 의 b 값과 비슷하다.
  - 그리고 $Activation func$ 를 통해, 조건이 만족하면 값을 출력한다.
  
  ----
  
  1. $x_1, x_2, x_3, \cdots , x_n$ $(feature)$
  2. $w_1, w_2, w_3, \cdots , w_n$ $(weight)$
  3. $xw \to x_1w_1, x_2w_2, \cdots , x_nw_n$
  4. $\begin{aligned} \sum_{k=1}^n{x_k\cdot w_k}\end{aligned}$
  5. $u = \begin{aligned} \sum_{k=1}^n{x_k\cdot w_k + b}\end{aligned}$
  6. $y = f(x) = f(\begin{aligned} \sum_{k=1}^n({x_k\cdot w_k) + b})\end{aligned}$
  

$$\vec{x} = (x_1\quad x_2 \quad x_3\cdots x_x)\\
\vec{w} = (w_1\quad w_2\quad w_3 \cdots w_n)\\
u = \vec{x} \cdot \vec{w} =  (x_1\quad x_2 \quad x_3\cdots x_x) \cdot 
\begin{pmatrix}
w_1\\
w_2\\
w_3\\
\vdots\\
w_n
\end{pmatrix} + b\\
$$

----

$$\\
n = XW+b\\
y = f(x) = f(XW+b)
$$


  - 딥러닝에서 가중치와 bias는 임의의 수로 설정 후 학습하는 과정에서 갱신함
  - 이번 예제는 가중치와 bias를 고정하고, 이 값들이 실제 신경망에서 어떤 의미인지를 확인
  - 입력으로 단 두개의 값(feature) x_1, x_2
  - 활성화함수로 시그모이드 함수 사용

```python
#x1, x2 값
X1 = np.arange(-1.0, 1.0, 0.2) # 원소 수는 10개
X2 = np.arange(-1.0, 1.0, 0.2)

#출력값을 저장할 10x10 그리드
Z = np.zeros((10,10))

#x1, x2 값의 입력 가중치
#w_x1 = 2.5
#w_x2 = 3.0
W = np.array([2.5,3.0])

#편향 --> 편향을 조정해서 신호 통과 여부를 임의로 변경할 수 있다.
bias = 0.1

#그리드맵의 각 그리드 별 뉴런의 연산
for i in range(10):
    for j in range(10):
        
        #입력과 가주이 곱의 합 + 편향
        list1 = [X1[i], X2[j]]
        X = np.array(list1)
        u = np.dot(X, W.T) + bias
        
          # u = X1[i]*w_x1 + X2[j]*w_x2+bias
        
        #그리드맵에 출력 값 저장
        y = 1/(1+np.exp(-u))#시그모이드 함수
        Z[j][i]=y
        
# 그리드 맵 표시
plt.imshow(Z, "gray", vmin=0.0, vmax = 1.0)
plt.colorbar()
plt.show()
```

$$
X = 
\begin{bmatrix}
x_1 & x_2 & x_3 & \cdots x_m
\end{bmatrix}
\quad feature\\
뉴런수 = n개\\
W=\begin{bmatrix}
w_{11} & w_{12} \cdots & w_{1n}\\
w_{21} & w_{22} \cdots & w_{2n}\\
\vdots & \vdots \qquad \vdots \\
w_{m1} & w_{m2} \cdots & w_{mn}
\end{bmatrix}\\
B = \begin{bmatrix}
b_1 & b_2 & b_3 & \cdots & b_n
\end{bmatrix}\\
XW = \begin{bmatrix}
\begin{aligned}
\sum_{k=1}^m{x_kw_{k1}} \quad & \sum_{k=1}^m{x_kw_{k2}} & \cdots \quad \sum_{k=1}^m{x_kw_{kn}}\\
\end{aligned}
\end{bmatrix}\\
U = \begin{bmatrix}
\begin{aligned}
\sum_{k=1}^m{(x_kw_{k1}) + b_1} \quad & \cdots \quad \sum_{k=1}^m{(x_kw_{kn}) + b_n}
\end{aligned}
\end{bmatrix}\\
Y = f(u) = \begin{bmatrix}\begin{aligned}f(\sum_{k=1}^m{(x_kw_{k1}) + b_1}),\quad \cdots \quad f(\sum_{k=1}^m{x_{k1}w_{kn}+b_n})\end{aligned}\end{bmatrix}
$$

### 아래의 코드 값은, 가중치의 형태에 따라 출력 값의 개수가 조정된다는 것을 보여 주기 위한 예시

```python
# 가중치의 형태에 따라 출력 값의 개수가 조정된다는 것을 보여 주기 위한 예시

#x1, x2 값
X = np.arange(-1.0, 1.0, 0.1) # 원소 수는 10개
Y = np.arange(-1.0, 1.0, 0.1)

#출력값을 저장할 10x10 그리드
Z = np.zeros((10,10))

#가중치
#w_im = np.array([[1.0,2.0],[2.0,3.0]]) #은닉층 2x2 행렬
#w_mo = np.array([[-1.0, 1.0],[1.0, -1.0]])#출력층 2x2 행렬


#편향 --> 편향을 조정해서 신호 통과 여부를 임의로 변경할 수 있다.
#b_im= np.array([0.3, -0.3]) #은닉층
#b_mo = np.array([0.4, 0.1]) #출력층


w_im = np.array([[2.0,2.0],[1.0,5.0]]) #은닉층 2x2 행렬
w_mo = np.array([[-2.0, 1.0],[-1.0, -1.0]])#출력층 2x2 행렬


b_im= np.array([-0.3, -0.3]) #은닉층
b_mo = np.array([0.4, -1.2]) #출력층



#은닉층
def middle_layer(x, w, b):
    u = np.dot(x,w) + b
    return 1/(1+np.exp(-u)) #시그모이드 함수

#출력층
def output_layer(x,w,b):
    u = np.dot(x,w) + b
    return np.exp(u)/np.sum(np.exp(u))

# 분류 결과를 저장하는 리스트
x_1 = []
y_1 = []
x_2 = []
y_2 = []
#그리드맵의 각 그리드 별 뉴런의 연산

for i in range(20):
    for j in range(20):
        
        # 순전파(forward propagation) : 딥러닝이 학습을 하면서 자신만의 답을 출력하는 과정.
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im) # 은닉층
        out = output_layer(mid, w_mo, b_mo) # 출력층
        
        #확률의 크기를 비교해 분류함
        if out[0] > out[1]:
            x_1.append(X[i])
            y_1.append(Y[j])
            
        else : 
            x_2.append(X[i])
            y_2.append(Y[j])
        
# 산포도 표시
plt.scatter(x_1, y_1, marker='+')
plt.scatter(x_2, y_2, marker='o')
plt.show()
```

## 역전파(back-propagaion)
  - 역전파는 신경망을 학습시킬 때 이용하는 알고리즘
  - 출력값과 정답의 오차를 네트워크에서 역전파시켜 네트워크의 가중치와 bias를 최적화시킴
  - <역전파 이해를 위한 항목>
    - 훈련 데이터, 테스트 데이터
    - 손실 함수
    - 경사 하강법
    - 최적화 알고리즘
    - 배치

### 1) 훈련 데이터, 테스트 데이터
  ※훈련 데이터와 테스트 데이터는 엄격히 구분한다※
  - 훈련 데이터 - 신경망이 학습에 이용되는 데이터
  - 테스트 데이터 - 학습 결과의 검증에 사용되는 데이터
  - 각 데이터는 여러 개의 입력값(feature)과 정답으로 구성
  
  - 입력값과 정답 한 쌍을 샘플이라 칭함
  - 일반적으로 훈련 데이터의 샘플수가 테스트 데이터의 샘플수보다 월등히 많음
  - 훈련 데이터로 학습이 제대로 된 신경망의 경우, 테스트 데이터에서 좋은 결과가 나옴

### 2) 손실함수 loss function / cost function 

  1. 평균제곱오차 - MSE(Mean squared error)
   - 회귀문제에 주로 사용
   
   $E = \frac {1}{n} \begin{aligned}\sum_{k=1}^n{(y_k - t_k)^2}\end{aligned}
   $

```python
def mean_squared_error(y,t):
    return np.sum(np.power(y-t, 2))/ y.shape[0]
```

  2. 오차제곱합 - SSE(Sum of Squares For Error)
    - 회귀문제에 주로 사용
    
   $E = \frac {1}{2} \begin{aligned}\sum_{k=1}^n{(y_k - t_k)^2}\end{aligned}
   $

```python
def square_sum(y,t):
    return 1.0 / 2.0 * np.sum(np.square(y-t))
```

확률이 골고루 높으면 엔트로피는 높다.
특정 정답일 확률이 높다 -> 엔트로피가 낮다

  3. 교차 엔트로피(Cross Entropy)
    - 분류 문제에 사용
    
       $E = - \begin{aligned}\sum_{k=1}^n{t_klog(y_k)}\end{aligned}$
       
    - 분류 문제에 사용
    
       $E = \begin{aligned}\sum_{k=1}^n{t_k(-log(y_k))}\end{aligned}
   $
   
    - 분류 문제에서 정답은 1이 하나이고 나머지는 모두 0인 원핫인코딩 벡터로 표현
    - 따라서 우변의 시그마 내부에서  $t_k$가 1인 항의 오차에만 영향을 줌
    - $y = -log(x)$ 의 그래프는 다음과 같음
      - import numpy as np
      
        import matplotlib.pyplot as plt
        
        x = np.linspace(start = 0.00001, stop=1.0, num=100)
        
        plt.plot(x, -np.log(x), 'b-', label="Natural Logarithm(ln)')
        
        plt.legend()
        
        plt.show()
    
    - x가 1일 때는 0이고, 0에 근접할수록 무한대로 커짐

    - 교차 엔트로피의 장점 중 하나는 출력값과 정답의 차이가 클 수록 학습 속도가 빨라짐

## 회귀 문제에서 출력층 기울기, 은닉층 기울기

### 1. 출력층 기울기
손실함수$E$ : 오치제곱합

은닉층 활성화함수$u_k$ : 시그모이드 

출력층 활성화 함수$y_k$ : 항등함수 

항등함수 이므로, $\frac {\partial y_k}{\partial u_k} = \frac {\partial u_k}{\partial u_k} = 1$
$$
\delta _k = \frac {\partial E}{\partial y_k} \cdot \frac{\partial y_k}{\partial u_k} \to \\
\frac {\partial E}{\partial y_k} = \frac {\partial}{\partial y_k}({\frac {1}{2}} \begin{aligned}\sum_{k=1}^n{(y_k-t_k)^2}\end{aligned}) = \frac {\partial}{\partial y_k} (\frac 12(y_1-t_1)^2 + \frac 12(y_2-t_2)^2 + \cdots + \frac 12(y_k-t_k)^2 +\cdots +\frac 12(y_n-t_n)^2) = y_k - t_k\\
※ \delta _k = y_k - t_k\\
$$

----
### 2. 은닉층 기울기
$$
\delta _j = \partial y_j \cdot \frac {\partial y_j}{\partial u_j} \to \\ 
\text {시그모이드의 도함수, } f'(x) = (1-f(x))f(x) 이므로 \quad \frac{\partial y_j}{\partial u_j} = (1-y_j)y_j \\
※\delta _j = \partial y_j \cdot (1-y_j)y_j
$$

  2. 모멘텀(Momentum)
    - SGD에 물리학적 관성 기법을 적용한 알고리즘
       - $W = W - \eta \frac {\partial E}{\partial W} +a\Delta w $
       - $b = b - \eta \frac {\partial E}{\partial b} +a\Delta b$
       - (최초의 모멘텀 $\Delta w와 \Delta b\text {의 값은 0})$
       - a는 모멘텀 상수(일반적으로 0.9사용)
      
       - 1) 장점
            - local minima에 잘 빠지지 않음

       - 2) 단점
            - 조정해야 하는 상수가 늘어남
            
  3. 아다그라드(Adaptive Gradient : AdaGrad)
    - 학습이 진행되면서 알고리즘에 의해 학습률이 조금씩 감소함
    $$
    h = h+(\frac {\partial E}{\partial w})^2\\
    w = w - \eta \frac 1{\sqrt{h}}\cdot \frac {\partial E}{\partial w}\\
    w = b - \eta \frac 1{\sqrt{h}}\cdot \frac {\partial E}{\partial b}
    $$
    
    - 첫 번째 식에서는 $h$는 학습이 진행되면서 계속 증가
    - 두 번째 식은 $h$가 분모에 있으므로 감소
    - $h$는 가중치별로 계산되므로, 그 때 까지의 총 수정량이 적은 가중치는 새로운 수정량이 증가, 또는 vice versa (반대로 작용)
    - 이런 로직 때문에 처음에는 넓은 영역에서 탐색을 시작해 점차 탐색범위를 좁혀가는 효율적인 탐색이 가능
    
    - 1) 장점
        - 조정해야할 하이퍼 파라미터가 $\eta$ 밖에 없음
        
    - 2) 단점
        - 수정량이 계쏙해서 감소하기 때문에 도중에 수정량이 0이 되어버려 더는 최적화가 진행되지 않을 수 있음 
        
  4. RMSProp
    - AdaGrad는 최솟값에 도달하기 전에 학습률이 0에 수렴할 수 있음
    - AdaGrad는 간단한 convex(볼록)함수에서는 잘 동작하지만, 복잡한 다차원 곡면 함수에서는 성능이 떨어짐. 즉 , 기울기의 단순한 누적만으로는 부족함
    $$
    h = \rho + (1+\rho )(\frac {\partial E}{\partial w})^2\\
    w = w - \eta \frac 1{\sqrt h} \frac {\partial E}{\partial w}
    $$
    - 일반적으로 $\rho$(로) = 0.9tkdyd
