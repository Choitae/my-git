# 빅 데이터 개념과 처리 과정

### 빅 데이터 등장 배경 
  - 1990년 이후 인터넷이 확산되면서 정영화된 데이터 외에도 비정형화 된, 무수히 발생하는 데이터를 분석할 필요성이 생김.
  - SNS 확산으로 정형화된 데이터로 처리가 제한, 그 때문에 비정형화된 데이터를 처리해야하는 필요성이 생김. 그러나 hardware, 인터넷의 한계로 발전이 크게 제한됨. 

### 빅데이터 개념과 처리과정
  - 데이터 양의 급속한 증가로 기존의 데이터 저장, 관리, 분석기법으로는 데이터 처리에 한계가 있어 정보 기술의 패러다임도 변화
  - PC시대 -> 인터넷 시대 -> 모바일 시대 -> IOT, Big Data -> 시마트 시대
  - 인터넷 시대에는 소프트웨어 판매와, 판매하는 것을 도와주는 플랫폼 제공 회사 발생 (ex) steam)
  - 모바일 시대에는 소프트웨어 판매를 스마트폰, 모바일 인터넷을 통해 좀 더 쉽고 편리하게 이용할 수 있음.
  - 시대가 변하면서 처리해야 할 데이터가 증가하고, 빅데이터 기반의 처리기술들이 많이 발생.
  - 4차산업은 마지막 , 종단에 데이터를 처리할 수 있는 어플 등을 달아 데이터를 받아 바로 사용가능할 수 있도록 실행하고자 한다.

### 빅데이터 개념과 속성
  - 규모
  - 다양성 : 정형화하기 어려운 데이터 포함한 다양한 데이터. 
  - 속도 : 데이터가 빠르게 생성되고 사라짐.
  - 정확성 : 데이터 표본이 워낙 많기 떄문에 오류를 줄여야 함.
  - 가치
  ※ 인공지능은 속도와 정확성을 향상 시키기 위하여 필수적으로 사용된다.

### 빅데이터 종류와 유형 변화
  - 정형 : 고정된 필드에 저장된 데이터. 관계형 데이터베이스, 스프레드시트 등
  - 비정형 : 고정된 필드에 저장되어 있지 않은 데이터. 이미지 동영상, 음성 데이터, 덱스트 분석이 가능한 텍스트 문서 등 

### 빅데이터 속성과 처리 특징
  - 다른 데이터 형식을, 대용량 데이터 규모로 받아 실시간 처리(스트림 데이터 폭증)
  - 신경제학, 새로운 데이터 소스 자료가 AI의 도움을 받아 빅데이터 사용
  - 데이터 소스 - 수집 - 저장 - 처리 - 분석 - 표현
  - 인공지능은 비정형화 된 데이터를 좋아하지 않기 때문에 저장된 데이터를 처리하는 과정(정형화시키는 작업)에 시간 소모가 많이 된다.
  ※ 데이터 수집 : 크롤링, ETL

### 빅테이터 소스 생성과 수집 기술
  - 로그 수집기, 크롤링, 세싱, RSS 리터/오픈 API를 이용하여 자료를 수집, ETL을 통해 데이터 추출,변환, 적재하여 데이터 Mart로 보냄.
  - NoSQL 몽고 data, Canssandra 등 여러 제품에서 데이터 저장.

### 빅데이터 처리 기술
  - 맵리듀스 : 키, 값 입력 데이터 분할 처리 및 처리 결과 통합 기술 (python)
  - R : R언어를 이용한 통계 기법, 모델링, 최신 데이터 마이닝 기법 등
  - 하둡 : 비정형화 데이터를 정형화된 데이터로 처리하는 실행 프레임워크

### 빅데이터 분석 기술
  - 마이닝 기법 : 많은 데이터 가운데 숨겨져 있는 유용한 상관관계를 발견하여, 미래에 실행 가능한 정보를 추출해 내고 의사 결정에 이용하는 과정을 말한다. - Naver 지식백과
  - 텍스트 마이닝 : 자연어 처리 기술을 이용하여 비정형 텍스트에서 유용한 정보를 추출하거나 다른 데이터와의 연계성을 파악, 분류, 군집화 등 빅데이터에 있는 의미 정보를 발견하는 것
  - 군집화 : 특성이 비슷한 데이터를 합쳐 군으로 분류하는 방법, 분류와 달리 훈련 데이터군을 이용하지 않기 떄문에 비지도 학습 사용
  - 비지도 학습? :  기계 학습의 일종으로, 데이터가 어떻게 구성되었는지를 알아내는 문제의 범주에 속한다. 이 방법은 지도 학습(Supervised Learning) 혹은 강화 학습(Reinforcement Learning)과는 달리 입력값에 대한 목표치가 주어지지 않는다. - 위키백과

### 아나콘다로 머신러닝 사용하기
  - 기본적으로 아나콘다에는 머신러닝에 필요한 프로그램들이 모두 설치가 되어있기 때문에, 아나콘다를 이용하여 수업을 진행할 것이다.

### 직선의 방정식을 이용한 머신러닝
  - y = Ax라는 직선의 방정식을 이용한다면.....
  - (1,3) / (3,1) 이라는 표본을 이용하여 사용하고, A의 시작값이 0.25로 한다면, 
  - (3,1)의 표본 아래로 직선이 지나가기 떄문에 사용할 수 없다 --> 값을 변화시켜야 한다.
  - y = 1.1로 하여 값을 구해본다면, E(오차) = 목표값 - 계산값 = 1.1 - 0.75 = 0.35
  - 목표과 근접한 값을 t라고하면, A값을 조정하여 얻는 값이 t이기 때문에,  t = (A +△A)x로 표기. 
  - E = t - y = y + △Ax - y = △Ax. △A = E/x --> 0.35 / 3.0 = 0.1167
  - y = (A +△A)x이면 현재 나온 값으로, y = (0.25 + 0.1167) X 3 = 1.1
  
  - 2번째로 (1,3)이라는 표본을 사용한다면,
  - y = (A +△A)x = (0.25 +8 0.1167) X 1.0 = 0.3367. 3과 너무 차이가 나기 때문에, 근사값 t = 2.9로 하면 오차 E는 2.9-0.3367 = 2.5333
  - △A = 2.5333 / 1.0 = 2.5333, 업데이트 된 (A +△A) = 2.9
  - 머신러닝에서 조정인자 L을 학습률 Learning Rate라고 하며 업데이트 방정식을(△A) = L(E/x)로 표현한다.

### 뉴런
  - 뉴런은 전기신호를 받아 전기신호로 출력한다. 즉, 전기신호로 소통한다. 
  - 뉴런 입력에 분계점에 도달해야 출력이 발생된다.
  
### 뉴런과 유사한 함수, '활성화 함수'
  - 활성화 함수 : 분계점을 넘어서는 출력신호를 생성하는 함수
  - 입력이 작은 경우 출력은 0, 분계점 이상이 되면 출력은 1
  - Sigmoid 함수가 대표적인 활성화함수. y = 1 / (1+(e)^(-1)(1+x))

### 인공 뉴런
  - 여러개의 뉴런이 연결된 것처럼, 여러개의 입력을 맏고 입력의 합을 구한 후, 시그모이드 함수를 통해 분계점을 넘게되면 출력. 아니면 미출력
  - 여러 계층과 계층에 설정된 활성화 함수로 뉴런으로 소통하는 것과 동일하게 만든다.
  - 뉴런간 연결의 신호 강도를 나타내는 가중치(weight)를 부여한다.
  - 활성화 함수는 가중치와 입력의 합이 입력되는데, x= aW_a + bW_b+ cW_c 형식이 된다.
