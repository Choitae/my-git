## 문제6

아래는 벡터 a의 Norm을 정의한 식이다. 




두 벡터를 내적하려면 다음과 같은 조건이 만족되어야 한다.

1.우선 두 벡터의 차원(길이)이 같아야 한다.
2.앞의 벡터가 행 벡터이고 뒤의 벡터가 열 벡터여야 한다

열벡터 x와 열벡터 y의 내적은 아래 같다.



위 2가지 식을 아래 셀에 마크다운드로 코딩하시오.

답:

$$
\text {벡터 a의 Norm :} \quad
\Vert a \Vert = \sqrt {a^Ta} = \sqrt {a_1^2+\cdots+a_N^2}
$$
$$
\text {열벡터 x와 열벡터 y의 내적 : } \quad
x^Ty=
\begin{bmatrix}
x_1&\quad x_2 & \quad \cdots& \quad x_N
\end{bmatrix}
\begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_N
\end{bmatrix}
=x_1y_1+\cdots+x_Ny_N
=\sum_{i=1}^N{x_iy_i}
$$
