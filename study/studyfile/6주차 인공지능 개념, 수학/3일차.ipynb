### 벡터와 백터의 곱셈
  - 벡터는 일반적으로 열 벡터이며, 내적을 하기 위해서는 전치를 통해 행 벡터로 바꾸어줘야한다.
  - 내적의 표기는 다음처럼 dot로 표기하는 경우도 있다.
$$
x \cdot y = \left < x, y \right > = x^Ty
$$
  - 두 벡터를 내적하려면 다음과 같은 조건이 만족되어야 한다.  
    1) 두 백터의 차원(길이)가 우선적으로 같아야 하며  
    2) 앞의 벡터가 행 벡터이고 뒤의 벡터가 열 벡터이어야 한다
$$
x^Ty = 
[x_1 x_2 \cdots x_N]
\begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_N
\end{bmatrix}
= x_1y_1+\cdots + x_Ny_N
= \sum_{i=1}^{N}x_iy_i
$$
  - 벡터의 내적 기호는 $@$ 또는 $dot()$를 사용한다.

#### 내적은 왜 할까?
  - 두 벡터간의 위치 관계를 알기 위하여, 내적은 보통 $u\cdot v\cdot cos\theta$ 형태로 나타내는데, 내적을 이용하여, 내적 값이 0이면 수직이다, 값이 음수이면 90도 이상으로 벌어져 있다, 등의 위치 정보를 알 수 있다.
  - 데이터 분석을 위해 사용된다. 가중합을 계산할 때 쓰일 수 있다.

### 가중합 
  - 벡터의 내적은 가중합을 계산할 때 자주 사용된다.
$$
x=[x_1,\cdots,x_n]^T, w=[w_1,\cdots,w_n]^T \text {일 때}\\
w_1x_1+\cdots+w_nx_n = \sum_{i=1}^{N}{w_ix_i}\\
\sum_{i=1}^{N}{w_ix_i}=
\begin{bmatrix}
w_1 & w_2 &\cdots & w_n \\
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2\\
\vdots\\
x_n
\end{bmatrix}
=w^Tx
=\begin{bmatrix}
x_1 & x_2 &\cdots & x_n \\
\end{bmatrix}
\begin{bmatrix}
w_1\\
w_2\\
\vdots\\
w_n
\end{bmatrix}
=x^Tw
$$

### 가중평균
  - 가중합의 가중치 값을 전체 가중치 값의 합으로 나누면 가중평균이 된다.
  - 예로, 주 1회 1시간 씩 하는 대학교 수업과 주 1회 3시간 씩 하는 대학교 수업의 성적이

### 행렬과 행렬의 곱셈
   - A행렬과 B행렬을 곱한 결과가 C행렬이 된다고 하자. C의 i번쨰 행, j번째 열의 원소 cij값은 A행렬의 i번째 행 벡터 $a_i^T$ 와 B 행렬의 j번째 열 벡터 $b_j$ 의 곱이다.
$$
C=AB\rightarrow c_{ij}=a_i^Tb_j
$$
  - 이 정의가 성립하려면 앞의 행렬 A열의 수가 뒤의 행렬 B의 행의 수와 일치해야만 한다.
$$
A\in R^{N\times L}, B\in  R^{L\times M} \rightarrow AB \in R^{N\times M}
$$
  - 예시
$$
A=
\begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6
\end{bmatrix}\\
B=
\begin{bmatrix}
1 & 2\\
3 & 4\\
5 & 6
\end{bmatrix}\\
C=AB=
\begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6
\end{bmatrix}
\begin{bmatrix}
1 & 2\\
3 & 4\\
5 & 6
\end{bmatrix}
=
\begin{bmatrix}
22 & 28\\
49 & 64
\end{bmatrix}
$$
   ※ 이러한 문제는 시험에 꼭 나온다!!!

### 여러 개의 벡터에 대한 가중합 동시 계산


#### 연습 문제 2.2.9
$$
\begin{bmatrix}
x_1^T\\
x_2^T
\end{bmatrix}
\begin{bmatrix}
w_1\\
w_2\\
w_3
\end{bmatrix}
$$

$$
\begin{bmatrix}
x_1^T\\
x_2^T
\end{bmatrix}
\begin{bmatrix}
w_1\\
w_2\\
w_3
\end{bmatrix}
=
\begin{bmatrix}
x_{11} & x_{21} & x_{32}\\
x_{12} & x_{22} & x_{32}
\end{bmatrix}
=
\begin{bmatrix}
x_1^Tw\\
x_2^Tw
\end{bmatrix}
$$

### 잔차
  - 선형 회귀분석을 한 결과는 가중치 벡터 $w$라는 형태로 나타나고,독립변수 데이터와 함께 벡터 $x_i$의 가중합인 $w^Tx_i$이 된다
  - 예측치와 실제값(target) $y_i$의 차이를 오차(error) 또는 잔차$e_i$라고 한다.

  - 모든 독립 변수 벡터에 대한 잔차값을 구하면 잔차 벡터 e가 된다.
  $$
e_i=y_i-\hat y_i = y_i - w^Tx_i
$$

$$
e=
\begin{bmatrix}
e_1\\
e_2\\
\vdots\\
e_M
\end{bmatrix}
=\begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_M
\end{bmatrix}
-\begin{bmatrix}
x_1^Tw\\
x_2^Tw\\
\vdots\\
x_n^Tw
\end{bmatrix}
$$

### 잔차 제곱합
  - 잔차의크기는 잔차 베거의 각 원소를 곱한 후 더한 잔차 제곱합(RSS)를 이용하여 구한다.
$$
\sum_{i=1}^Ne_i^2=\sum_{i=1}^N(y_i-w^Tx_i)^2=\sum_{i=1}^N{(e_i*e_i)}=e^Te=(y-Xw)^T(y-Xw)
$$


$$
(y-Xw)^T(y-Xw)=y^Ty-w^TX^Ty-y^TXw+w^TX^TXw
$$

$$
(y-Xw)^T=y^T-w^TX^T\text{이므로}\\
(y-Xw)^T(y-Xw)=(y^T-w^TX^T)(y-Xw)=y^Ty-w^TX^Ty-y^TXw+w^TX^TXw
$$

### 이차형식
  - 전차 제곱합의 마지막 항은 $w^TX^TXw$로 $X^TX$는 정방행렬이 된다. 이를 A라고 하면 마지막 항은 $w^TAw$와 같은 형태가 된다. 이는 행 벡터 * 정방행렬 * 열 백터  형식이다.
  

## 부분행렬
  - 일반적인 2차원 정방행렬 A,B의 곱을 보자
  - 우선 앞에 곱해지는 행렬을 행벹터로 나누어 계산해도 된다.
  - 아니면 뒤에 곱해지는 행렬을 열벡터로 나누어 계산해도 된다.
  - 앞에 곱해지는 행렬을 열벡터로, 뒤에 곱해지는 행렬을 행벡터로 나누어 스칼라처럼 계산해도 된다.
