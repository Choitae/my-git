```python
snow_storm_dataframe_columns = ['Lat', 'Lon', 'snowfall', 'STORM_ID', 'START_DATE', 'END_DATE']
```

```python
snow_storm_refined_data = pd.DataFrame(columns=snow_storm_dataframe_columns)
```

```python
snow_storm_refined_data = snow_storm_refined_data.astype({'snowfall':float, 'Lat':float, 'Lon':float})
```

```python
snow_storm_refined_data.info()
```

```python
snow_storm_refined_data
```

```python
for dataframe in tqdm(snow_storm_dataframe_list):
    temp = dataframe[snow_storm_dataframe_columns]
    snow_storm_refined_data = snow_storm_refined_data.append(temp, ignore_index=True)
```

```python
snow_storm_refined_data.info()
```

```python
snow_storm_refined_data
```

```python
snow_storm_refined_data.info()
```

```python
snow_storm_refined_data['start_year'] = snow_storm_refined_data['START_DATE'].astype(str).str.slice(start=0, stop=4).astype(int)
```

```python
snow_storm_refined_data['START_DATE'] = snow_storm_refined_data['START_DATE'].astype('datetime64')
```

```python
snow_storm_refined_data['END_DATE'] = snow_storm_refined_data['END_DATE'].astype('datetime64')
```

```python
snow_storm_refined_data['duration'] = snow_storm_refined_data['END_DATE'] - snow_storm_refined_data['START_DATE']
```

```python
snow_storm_refined_data['duration'] = snow_storm_refined_data['duration'].astype(str).str.slice(start=0, stop=-4).astype(int)
```

```python
snow_storm_refined_data['count'] = 1
```

```python
plt.scatter(snow_storm_refined_data.groupby('STORM_ID').max()['start_year'], snow_storm_refined_data.groupby('STORM_ID').max()['snowfall'], s=1)
plt.figure(figsize=(30,20))
plt.show()
```

```python
fql = np.polyfit(snow_storm_refined_data.groupby('STORM_ID').max()['start_year'], snow_storm_refined_data.groupby('STORM_ID').max()['snowfall'],1)
f1= np.poly1d(fql)
plt.scatter(snow_storm_refined_data.groupby('STORM_ID').max()['start_year'], snow_storm_refined_data.groupby('STORM_ID').max()['snowfall'], s=1)
plt.plot(snow_storm_refined_data.groupby('STORM_ID').max()['start_year'],f1(snow_storm_refined_data.groupby('STORM_ID').max()['start_year']), color = 'red')
plt.figure(figsize=(12,8))
plt.show()
#연도별 발생하는 눈폭풍 들의 최대 적설량을 표현
```

```python
fql = np.polyfit(snow_storm_refined_data.groupby('STORM_ID').mean()['start_year'], snow_storm_refined_data.groupby('STORM_ID').mean()['snowfall'],1)
f1= np.poly1d(fql)
plt.scatter(snow_storm_refined_data.groupby('STORM_ID').mean()['start_year'], snow_storm_refined_data.groupby('STORM_ID').mean()['snowfall'], s=1)
plt.plot(snow_storm_refined_data.groupby('STORM_ID').mean()['start_year'],f1(snow_storm_refined_data.groupby('STORM_ID').mean()['start_year']), color = 'red')
plt.figure(figsize=(12,8))
plt.show()
# 연도별 발생하는 눈폭풍 들의 평균 적설량




```
