```python
def sum_data_temp(data,year_data,month_data):
    find_list = []
    for year in year_data:
        sum_data = 0
        for month in month_data:    
            find_date = year + '-' + month + '-01'
            sum_data += round(float(data[data.index==find_date]['AverageTemperature']),2)
        find_list.append(sum_data)
    return find_list
```

```python
def arg_winter_data_temp(data,year_data,month_data):
    find_list = []
    for year in year_data:
        sum_data = 0
        for month in month_data:
            if month == '12':
                find_date = str(year) + '-' + month + '-01'
                sum_data += round(float(data[data.index==find_date]['AverageTemperature']),2)   
            else : 
                find_date = str(year+1) + '-' + month + '-01'
                sum_data += round(float(data[data.index==find_date]['AverageTemperature']),2)
        find_list.append(round(sum_data/3 , 2))
    return find_list
```

```python
def arg_summer_data_temp(data,year_data,month_data):
    find_list = []
    for year in year_data:
        sum_data = 0
        for month in month_data:    
            find_date = str(year) + '-' + month + '-01'
            sum_data += round(float(data[data.index==find_date]['AverageTemperature']),2)
        find_list.append(sum_data/3)
    return find_list
```

```python
def wint_findfunc(db, year):
    return_list = []
    month = ['12', '01', '02']
    for y in year :
        
        date_list= []
        data_sum=0
        for m in month :
            if m == '12':
                date = str(y) + '-' + m
                data_sum += findfunc(db,date)
            else : 
                date = str(y+1) +'-' +m
                data_sum += findfunc(db,date)
        return_list.append(round(data_sum/3,4))
    return return_list
```

```python
def summ_findfunc(db, year):
    return_list = []
    month = ['06', '07', '08']
    for y in year :
        date_list= []
        data_sum=0
        for m in month :
            date = str(y) + '-' + m
            data_sum += findfunc(db,date)
        return_list.append(round(data_sum/3,4))
    return return_list
```
