### SQL

데이터베이스-수학의 관계 배수를 통해 저장됨

여러명의 사용자가가 사용할 수 있도록 응용프로그램이 공유 .

데이터베이스는 
1) 전체를 관리하며 사용자에게 일정부분의 공간을 부여해주는 방법
2) 새로운 데이터를 추가할 때마다 하드디스크를 추가하는 것과 비슷하게 데이터베이스 마다 별도의 공간을 생성시켜주고 그곳에 저장하는 방법.
으로 크게 두 가지로 나눌 수 있다.

접속 방법으로는 
1) 데이터베이스에 직접 접속하는 방법. ip와 port가 필요함.
2) Connector interface라는 각 프로그래밍 언어에 맞는 명세서와 비슷한 것을 만들어 
틀에 맞추어 접속하는, 데이터베이스에 접속하지 않고도 수행이 가능한 방법.

사용방법으로는 SQL을 이용한다. 관계 배수를 통해 프로그래밍이 이루어지며 중복되는 데이터를 좋아하지 않음.


데이터베이스 또는 DBMS의 특징
1) 데이터의 무결성
   - 어떤 경로를 통해 들어왔든 데이터에 오류가 없어야 한다.
     --> 제약 조건을 통해 무결성을 구현한다. (Table을 구성)
          -- 반드시 있어야 하는 데이터
          -- 중복되지 않는 데이터(중복의 최소화) 
2) 보안
   - 접근이 허가된 사람만이 데이터에 접근 가능.
3) 데이터의 안정성 향상
   - 사용 시 다른 사람이 접속하는 것을 방지하는 등 트렌젝션의 실현성 보장
   - 데이터 손실 시 복원/복구
4) 데이터의 독립성
   - 어떤 프로그래밍 언어를 사용하던지, 어떤 위치에서 접속하던지 이상없이 작동 될 수 있도록.
   - 데이터 외부단계(사용자) - 개념단계 - 내부단계(실질적으로 DB관리자가 사용하는 부분)
5) 데이터 중복의 최소화
   - 하나의 테이블로 저장 / 관리
     --> 통일된 방식으로 응용 프로그램 작성 가능, 유지보수도 쉬움.

데이터 베이스의 발전
--> 오프라인 (종이)에서 파일(데이터)로 변경
--> 데이터베이스 관리 시스템(DBMS)
      - 대량의 데이터를 보다 효율적으로 관리, 운영하기 위해 사용
      - 파일시스템의 단점 보완(가격면, 관리 효율성 등 대부분의 요소에서 파일 개개로 관리하는 것보다 우수)
      - SQL 데이터를 구축하고 관리, 활용하기 위한 언어

DBMS 분류
1) 계층형 DBMS
   - 각 계층은 트리 Tree 형태 - 1:N 관계. 1960년대에 시작
   - 처음 구축한 이후 구조 변경이 까다로움
   - 접근의 유연성이 부족. 주어진 조건에서는 매우 빠른 검색이 되나 임의의 조건에서는 효율이 낮음
2) 망형 DBMS
   - 1:1, 1:N, N:M (다대다) 관계 지원
   - 효과적이고 빠른 데이터 추출 가능 
   - 복잡한 내부 포인터로 프로그래머가 이 모든 구조를 이해해야만 프로그램의 작성이 가능하다는 단점을 가짐.
3) 관계형 DBMS(RDBMS)
   - E.F.Codd의 수학 모델에 근거해 고안
   - 키와 값들의 간단한 관계를 테이블화 시기킨 매우 간단한 원칙 개념의 데이터베이스
   - 테이블이라 불리는 최소 단위로 구성. 테이블은 하나 이상의 열로 구성

## python DB 연동

## 출력하기

- 데이터베이스 연결 --> 연결자 = sqlite3.connect("DB이름")
  - 커서 생성 --> 커서이름 = 연결자.cursor()
  - 데이터 조회 --> 커서이름.execute("SELECT 문장") #select *(모든 데이터) from 테이블 이름(꺼낼 위치)
  - 조회한 데이터 출력[반복] --> 커서이름.fetchone()
  - 데이터베이스 닫기 --> 연결자.close()

## 삽입하기
- 데이터베이스 연결 --> 연결자 = sqlite3.connect("DB이름")
  - 커서 생성 --> 커서이름 = 연결자.cursor()
  - (테이블 만들기) --> 커서이름.execute("CREATE TABLE 문장") 
  - 데이터 입력 --> 커서이름.execute("INSERT 문장")
  - 입력한 데이터 저장 --> 연결자.commit()
  - 데이터베이스 닫기 --> 연결자.close()

## ORM
  - 객체 관계 매핑
    - 객체와 관계형 데이터베이스(RDBMS)를 매핑(Mapping)한다.
    - 객체 지향 프로그래밍(OOP)과 관계형 데이터베이스와(RDBMS)는 구성이 다름.
      객체 중심의 프로그래밍과 달리 데이터베이스는 데이터중심(테이블)으로 구조화 목적. 일련의 정형화된 테이블로 구성된 데이터 항목들의 집합체. 때문에 OOP와 RDBMS는 지향하는 목적이 서로 달라 기능과 표현 방법이 다르다.
