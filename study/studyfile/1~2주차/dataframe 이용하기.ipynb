```python
df = pd.DataFrame([['세이 두바이', 2020, '홍길동', '이순신', 300, 100.0, 10]], columns=df_bond.columns) #2차원 배열로 만들어야 한다.
df
```

### append()

```python
df_bond.append(df).tail()
```

```python
df_bond.append(df, ignore_index=True)
```

```python
df_bond = df_bond.append(df, ignore_index=True)
df_bond.tail(3)
```

### concat()

```python
df = pd.DataFrame([['세이 두바이2', 2020, '홍길동1', '이순신1', 300, 100.0, 10]], columns=df_bond.columns) #2차원 배열로 만들어야 한다.
df
```
