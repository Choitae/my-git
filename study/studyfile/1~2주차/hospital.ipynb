# 1. 전체 병원 / 주소 / 의사수 목록

```python
def all_hosp(a) : 
    print("전체병원 / 주소 / 의사수")
    hops_list = []
    for i in a :
        hops_list += i[0], i[3], i[6]
    return hops_list
```

# 2. 과목명 병원 검색 출력

```python
def field_hosp(a):
    field = input("원하는 병원 분야 병원 : ")
    hops_list = []
    for i in a :
        if i[1] == field:
            hops_list += i
    return hops_list
```
