## Git과 GitHub_Desktop 연동 이후부터 설명

- branch를 하나 생성
  - master을 이용하게 되면, 파일을 불필요하게 수정하거나 복구하는 일이 일어나기 때문에, branch를 하나 생성하여, master의 파일을 복사하고, 이를 이용해서 작업, 이후 관리자 승인을 통한 합쳐주기를 실행하기.
  - 기본적으로 base를 master로 선택하여 생성하면 된다.




- Publish bracnh 클릭
  - git 홈페이지와 desktop git을 연동.
 

- Open in Visual Studio Code 
  - 여기서 작업 실시. 파일 생성, 수정, 제거 후 저장해준다.
  
  

- 아래 그림 빈칸 입력 후 Commit 실행

- Push Origin 클릭
  - Git hub에 연동 시켜준다.
- Create Pull Request 클릭
  - Git hub의 페이지가 오픈된다.


- Create Pull Request 클릭


- Merge pull Request 클릭
 

- Confirm Merge 클릭


- 완료
  - Git 홈페이지에 들어가면, 내용이 수정되어있음을 볼 수 있음.

- 판단하여, branch가 불필요하다고 판단 시 제거.
  - 그러나, Github_Desktop에는 branch가 남아있다. 별도로 제거할 것.
